# Derivation of Logistic Log Loss Function

## Step 1: Logistic Regression Model

The logistic regression model predicts probabilities for binary classification. The predicted probability for the positive class is:

The logistic regression model predicts probabilities for binary classification. The predicted probability for the positive class is:

$
\hat{y} = P(y=1 \mid x) = \frac{1}{1 + e^{-z}}
$

where:
- $ z = w^\top x + b $ is the linear combination of inputs and weights.
- $ \hat{y} $ is the predicted probability for the positive class ($ y=1 $).

The probability for the negative class ($ y=0 $) is:

$
P(y=0 \mid x) = 1 - \hat{y}
$

---

## Step 2: Likelihood Function

The likelihood of observing the data is:

$
L(w, b) = \prod_{i=1}^N P(y_i \mid x_i)
$

Substituting the probabilities:

$
P(y_i \mid x_i) = \hat{y}_i^{y_i} (1 - \hat{y}_i)^{1 - y_i}
$

The likelihood becomes:

$
L(w, b) = \prod_{i=1}^N \hat{y}_i^{y_i} (1 - \hat{y}_i)^{1 - y_i}
$

---

## Step 3: Log-Likelihood Function

Taking the logarithm of the likelihood (log-likelihood):

$
\ell(w, b) = \log L(w, b) = \sum_{i=1}^N \log \left( \hat{y}_i^{y_i} (1 - \hat{y}_i)^{1 - y_i} \right)
$

Using the property of logarithms:

$
\ell(w, b) = \sum_{i=1}^N \left[ y_i \log(\hat{y}_i) + (1 - y_i) \log(1 - \hat{y}_i) \right]
$

---

## Step 4: Negative Log-Likelihood (Log Loss)

To minimize the loss, we take the negative of the log-likelihood:

$
\text{Log Loss} = -\ell(w, b)
$

Substituting the expression for $\ell(w, b)$:

$
\text{Log Loss} = -\sum_{i=1}^N \left[ y_i \log(\hat{y}_i) + (1 - y_i) \log(1 - \hat{y}_i) \right]
$

---

## Step 5: Final Log Loss Function

Taking the average over all samples:

$
\text{Log Loss} = -\frac{1}{N} \sum_{i=1}^N \left[ y_i \log(\hat{y}_i) + (1 - y_i) \log(1 - \hat{y}_i) \right]
$

where:
- $ y_i $ is the true label (0 or 1).
- $ \hat{y}_i $ is the predicted probability for $ y_i = 1 $.
- $ N $ is the total number of samples.

---

## Key Intuition

For each prediction:
- If $ y=1 $, the loss is $-\log(\hat{y})$. Higher $\hat{y}$ (closer to 1) reduces the loss.
- If $ y=0 $, the loss is $-\log(1-\hat{y})$. Lower $\hat{y}$ (closer to 0) reduces the loss.

# Train the Logistic Regression with SGD

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import seaborn as sns
import pandas as pd

# Generate random dummy dataset
np.random.seed(42)
n_samples = 17
X = np.random.randn(n_samples, 2)
y = (X[:, 0] + X[:, 1] > 0).astype(int)

# Create DataFrame
X_df = pd.DataFrame(X, columns=['Feature 1', 'Feature 2'])
X_df['Target'] = y

print("Dummy Dataset:")
display(X_df)

# Sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Logistic Regression with SGD
class LogisticRegression:
    def __init__(self, learning_rate=0.1):
        self.learning_rate = learning_rate
        self.weights = None
        self.bias = None
        self.losses = []

    def initialize_parameters(self, n_features):
        self.weights = np.zeros(n_features)
        self.bias = 0

    def fit(self, X, y, iterations):
        self.initialize_parameters(X.shape[1])

        for i in range(iterations):
            # SGD: randomly select one sample
            idx = np.random.randint(0, X.shape[0])
            x_i = X[idx]
            y_i = y[idx]

            # Forward pass
            z = np.dot(x_i, self.weights) + self.bias
            y_pred = sigmoid(z)

            # Gradient computation
            dw = x_i * (y_pred - y_i)
            db = y_pred - y_i

            # Update parameters
            self.weights -= self.learning_rate * dw
            self.bias -= self.learning_rate * db

            # Store loss
            self.losses.append(self._compute_loss(X, y))

    def predict(self, X):
        z = np.dot(X, self.weights) + self.bias
        return (sigmoid(z) >= 0.5).astype(int)

    def _compute_loss(self, X, y):
        z = np.dot(X, self.weights) + self.bias
        y_pred = sigmoid(z)
        return -np.mean(y * np.log(y_pred + 1e-10) + (1 - y) * np.log(1 - y_pred + 1e-10))

# Train and evaluate model for different iterations
iterations_list = [10, 50, 100]
metrics = {
    'accuracy': [],
    'precision': [],
    'recall': [],
    'f1': [],
    'confusion_matrices': []
}

plt.figure(figsize=(15, 10))

for i, n_iter in enumerate(iterations_list):
    # Train model
    model = LogisticRegression(learning_rate=0.1)
    model.fit(X, y, n_iter)
    y_pred = model.predict(X)

    # Calculate metrics
    metrics['accuracy'].append(accuracy_score(y, y_pred))
    metrics['precision'].append(precision_score(y, y_pred))
    metrics['recall'].append(recall_score(y, y_pred))
    metrics['f1'].append(f1_score(y, y_pred))
    metrics['confusion_matrices'].append(confusion_matrix(y, y_pred))

    # Plot decision boundary with new colors
    plt.subplot(2, 3, i+1)
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                         np.arange(y_min, y_max, 0.1))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.contourf(xx, yy, Z, alpha=0.3)
    scatter = plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.RdYlBu, alpha=0.9)
    plt.title(f'Decision Boundary ({n_iter} iterations)')

    # Plot loss curve
    plt.subplot(2, 3, i+4)
    plt.plot(model.losses)
    plt.title(f'Loss Curve ({n_iter} iterations)')
    plt.xlabel('Iteration')
    plt.ylabel('Loss')

plt.tight_layout()

# Print metrics
print("\nPerformance Metrics:")
for i, n_iter in enumerate(iterations_list):
    print(f"\nIterations: {n_iter}")
    print(f"Accuracy: {metrics['accuracy'][i]:.3f}")
    print(f"Precision: {metrics['precision'][i]:.3f}")
    print(f"Recall: {metrics['recall'][i]:.3f}")
    print(f"F1-score: {metrics['f1'][i]:.3f}")
    print("\nConfusion Matrix:")
    print(metrics['confusion_matrices'][i])
```

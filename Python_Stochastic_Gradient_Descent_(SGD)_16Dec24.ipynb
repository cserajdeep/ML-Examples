# Stochastic Gradient Descent: Mathematical Foundations

## 1. Objective Function J(θ)

The objective function in machine learning is typically formulated as the average loss across all training examples:

$J(θ) = \frac{1}{n} \sum_{i=1}^{n} L(h_θ(x^{(i)}), y^{(i)})$

Where:
- $J(θ)$ is the cost function
- $n$ is the number of training examples
- $L$ is the loss function (e.g., mean squared error, cross-entropy)
- $h_θ(x)$ is the hypothesis/prediction function
- $x^{(i)}$ is the $i$-th training example
- $y^{(i)}$ is the corresponding true label

## 2. Gradient Computation

In Stochastic Gradient Descent (SGD), we compute the gradient of the objective function:

$∇J(θ) = \frac{1}{n} \sum_{i=1}^{n} \nabla L(h_θ(x^{(i)}), y^{(i)})$

The key difference from batch gradient descent is that SGD typically uses only a single training example (or a small batch) to estimate the gradient, introducing stochasticity.

## 3. Parameter Update Rule

The core update mechanism in SGD is:

$θ_{t+1} = θ_t - η ∇J(θ_t)$

Where:
- $θ_t$ is the parameter vector at time step $t$
- $η$ (eta) is the learning rate
- $∇J(θ_t)$ is the gradient at the current parameter values

## 4. Stochastic Gradient Descent Implementation Considerations

1. **Learning Rate**: Crucial hyperparameter that determines step size
   - Too small: Slow convergence
   - Too large: Might overshoot optimal solution

2. **Variations**:
   - Mini-batch SGD: Uses a small random subset of data
   - Momentum SGD: Adds accumulated gradient information
   - Adaptive methods: Adam, RMSprop adjust learning rates

## 5. Python Example Skeleton

```python
def sgd_update(params, gradients, learning_rate):
    """
    Perform SGD parameter update
    
    Args:
    - params: Current parameter values
    - gradients: Computed gradients
    - learning_rate: Step size for updates
    
    Returns:
    - Updated parameters
    """
    for param, grad in zip(params, gradients):
        param -= learning_rate * grad
    return params
```

## Mathematical Insights

- SGD introduces noise in gradient estimation
- This noise can help escape local minima
- Provides regularization effect
- Computationally more efficient than batch gradient descent

````python
import numpy as np
import random
import matplotlib.pyplot as plt

# Define the objective function and its gradient
def objective_function(x):
    return (x - 2)**2  # A parabola with a minimum at x=2

def gradient_function(x):
    return 2 * (x - 2)  # Gradient of the objective function

# SGD implementation
def stochastic_gradient_descent(learning_rate=0.1, epochs=100, initial_x=None):
    if initial_x is None:
        # Initialize x randomly in the range [-10, 10]
        x = random.uniform(-10, 10)
    else:
        x = initial_x

    loss_history = []  # To store the loss values for plotting

    for epoch in range(epochs):
        # Compute the gradient
        gradient = gradient_function(x)

        # Update x using the SGD update rule
        x = x - learning_rate * gradient

        # Compute and store the loss
        loss = objective_function(x)
        loss_history.append(loss)

        # Print the progress
        print(f"Epoch {epoch + 1}: x = {x:.4f}, f(x) = {loss:.4f}")

    # Plot the loss over iterations
    plt.plot(range(1, epochs + 1), loss_history, marker='o')
    plt.title('Loss Reduction Over Iterations')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.grid(True)
    plt.show()

    return x

# Run the SGD optimizer
optimized_x = stochastic_gradient_descent(learning_rate=0.1, epochs=50)
print(f"\nOptimized x: {optimized_x:.4f}")
````

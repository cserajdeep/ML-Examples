```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Define the RBF function
def rbf(x, c, gamma):
    return np.exp(-gamma * (x - c)**2)

# Define the RBF neural network
class RBFNN:
    def __init__(self, num_centers, gamma):
        self.num_centers = num_centers
        self.gamma = gamma
        self.centers = np.random.uniform(-np.pi, np.pi, num_centers)
        self.weights = np.random.uniform(-1, 1, num_centers)

    def fit(self, X, y):
        self.centers = X[np.random.choice(X.shape[0], self.num_centers, replace=False)]
        self.weights = np.linalg.pinv(self.rbf_matrix(X, self.centers, self.gamma)).dot(y)

    def rbf_matrix(self, X, centers, gamma):
        return np.array([rbf(X, c, gamma) for c in centers]).T

    def predict(self, X):
        return self.rbf_matrix(X, self.centers, self.gamma).dot(self.weights)

# Generate training data
X_train = np.linspace(-np.pi, np.pi, 100)
y_train = np.sin(X_train)

# Train the RBFNN
rbfnn = RBFNN(num_centers=10, gamma=1)
rbfnn.fit(X_train, y_train)

# Generate test data
X_test = np.linspace(-np.pi, np.pi, 1000)
y_test = np.sin(X_test)

# Predict using RBFNN
y_pred = rbfnn.predict(X_test)

# Evaluate the model
mse_test = mean_squared_error(y_test, y_pred)
mae_test = mean_absolute_error(y_test, y_pred)
rmse_test = np.sqrt(mse_test)
r2_test = r2_score(y_test, y_pred)

print("Test Metrics:")
print(f"MSE: {mse_test:.4f}, MAE: {mae_test:.4f}, RMSE: {rmse_test:.4f}, RÂ²: {r2_test:.4f}")

# Plot the results
plt.plot(X_test, y_test, label='sin(x)')
plt.plot(X_test, y_pred, label='RBFNN approximation')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Define the RBF function
def rbf(x, c, gamma):
    return np.exp(-gamma * (x - c)**2)

# Define the RBF neural network
class RBFNN:
    def __init__(self, num_centers, gamma):
        self.num_centers = num_centers
        self.gamma = gamma
        self.centers = np.random.uniform(-np.pi, np.pi, num_centers)
        self.weights = np.random.uniform(-1, 1, num_centers)

    def fit(self, X, y):
        self.centers = X[np.random.choice(X.shape[0], self.num_centers, replace=False)]
        self.weights = np.linalg.pinv(self.rbf_matrix(X, self.centers, self.gamma)).dot(y)

    def rbf_matrix(self, X, centers, gamma):
        return np.array([rbf(X, c, gamma) for c in centers]).T

    def predict(self, X):
        return self.rbf_matrix(X, self.centers, self.gamma).dot(self.weights)

# Generate training data
X = np.linspace(-np.pi, np.pi, 200)
y = np.sin(X)

# Split data into training and test sets
X_train, X_test = X[:150], X[150:]
y_train, y_test = y[:150], y[150:]

# Train the RBFNN
rbfnn = RBFNN(num_centers=10, gamma=1)
rbfnn.fit(X_train, y_train)

# Predict using RBFNN
y_pred_train = rbfnn.predict(X_train)
y_pred_test = rbfnn.predict(X_test)

# Evaluate the model
mse_train = mean_squared_error(y_train, y_pred_train)
mae_train = mean_absolute_error(y_train, y_pred_train)
rmse_train = np.sqrt(mse_train)
r2_train = r2_score(y_train, y_pred_train)

mse_test = mean_squared_error(y_test, y_pred_test)
mae_test = mean_absolute_error(y_test, y_pred_test)
rmse_test = np.sqrt(mse_test)
r2_test = r2_score(y_test, y_pred_test)

print("Training Metrics:")
print(f"MSE: {mse_train:.4f}, MAE: {mae_train:.4f}, RMSE: {rmse_train:.4f}, RÂ²: {r2_train:.4f}")
print("Test Metrics:")
print(f"MSE: {mse_test:.4f}, MAE: {mae_test:.4f}, RMSE: {rmse_test:.4f}, RÂ²: {r2_test:.4f}")

# Plot the results
plt.plot(X_train, y_train, 'bo', label='Training Set')  # blue circles for training set
plt.plot(X_test, y_test, 'go', label='Test Set')  # green circles for test set
plt.plot(X_train, y_pred_train, 'r-', label='Predicted Train')  # red line for predicted train
plt.plot(X_test, y_pred_test, 'r--', label='Predicted Test')  # red dashed line for predicted test
plt.legend()
plt.show()
```
